# Demonstration - Performance metrics

Pieter Overdevest  
2022-09-18

For evaluation of classification models, commonly used performance
metrics include, precison, recall, and false positive rate \[1\]. The
latter two make up the Receiver Operating Curve (ROC) \[2\]. Another
important curve - specially in case the data is imbalanced, as we will
show here - is the Precision/Recall Curve \[3\].

[This app](https://innovatewithdata.shinyapps.io/performance-metrics/)
demonstrates the influence of three aspects on the ROC and
Precision/Recall curve for a logistic regression model. The aspects
include, (1) the threshold value (k), (2) imbalanced data, and (3) the
level of separation between the positive and negative populations in the
feature space. Here, the feature space is constructed from a single
feature (x) to simplify visualisation. For the example above, blood
pressure is the single feature, the positive population got a heart
disease, while the negative population did not get a heart disease.
