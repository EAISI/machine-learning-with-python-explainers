# Demonstration - Quality of fit

Pieter Overdevest  
2022-09-18

The main purpose of machine learning models is to make them work well on
unseen data. Of course, using the available data you can tweak and add
more flexibility to the model untill its error approaches zero. However,
you may be disappointed when applying the model on unseen data. When
this happens, this phenomenon is known as ‘overfitting’, which
essentially means that the model follows the noise in the data too
closely. On the other hand, too little flexibility restricts the model
in its ability to follow the signal in the data. These observations are
caused by two competing properties of statistical learning methods: bias
and variance. The relation between these two is referred to as the
‘bias-variance trade-off’.\[1\]

[This app](https://innovatewithdata.shinyapps.io/quality-of-fit/)
demonstrates these phenomena by iteratively fitting polynomial or spline
models - with different levels of flexibility - on training sets
re-sampled from the same simulated data. The train/test split is 70/30.
The trained model is evaluated using both the training and the test
sets, blue and red, resp.
