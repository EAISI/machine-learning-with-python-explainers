## Tips

1.  Comment-out code:

-   On Windows, the shortcut is: CTRL + /
-   On Mac, the shortcut is: Command + /

1.  Download single files from GitHub, e.g., data file:
    https://downgit.github.io

2.  Add ‘;’ at the end of a code line to create a plot. This prevents
    statements like ‘\<AxesSubplot:\>’.

3.  When you use that same/similar code twice consider a function; use
    the same/similar code three times or more, create a function.

4.  Use ’' as a carriage return to continue the script on the next line.
    This keeps code readable and you don’t have to put everything in one
    line.

## Best practices

### Object naming convention in Intermezzo’s and Ames step-by-step

In the intermezzo’s and in ‘Ames step-by-step’ the object names are
prepended to clarify the data type and to make it easier to find object
names in environments running *intellisense*.

**Scalar**

-   Number: `n_`

-   Character: `c_`

-   Boolean: `b_`

**One-dimensional**

-   Pandas series: `ps_`

-   List: `l_`

-   Array (1D): `v_`

-   Dictionary: `dc_`

**Two-dimensional**

-   Data frame: `df_`

-   Array (2D): `m_` (matrix), `sm_` (sparse matrix)

**Other**

-   Function: `f_`

-   Pipeline: `pl_`

-   Array of higher dimension: `ar_`

-   Model: `mo_`
